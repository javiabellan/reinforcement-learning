# Evaluate

```python
def score(my_agent, advesary, num_episodes):
    rewards     = evaluate("connectx", [my_agent, advesary], num_episodes=num_episodes)
    mean_reward =  sum(r[0] for r in rewards) / sum(r[0] + r[1] for r in rewards)
    print("My Agent vs", advesary, " gets ", mean_reward*100, "%")

          
score(random_agent, "random",  10)
score(random_agent, "negamax", 10)
```

# Convert it to gym

```python
class ConnectX(gym.Env):

    def __init__(self):
        self.env = make("connectx", debug=True)
        self.trainer = self.env.train([None, "random"])

        n_columns, n_rows      = self.env.configuration['columns'], self.env.configuration['rows']
        self.action_space      = gym.spaces.Discrete(n_columns)
        #self.observation_space = gym.spaces.Box(low=0, high=2, shape=(n_columns, n_rows))
        self.observation_space = gym.spaces.Discrete(n_columns * n_rows)

    def step(self, action):
        return self.trainer.step(action)
    
    def reset(self):
        return self.trainer.reset()
    
    def render(self, **kwargs):
        return self.env.render(**kwargs)
    
env = ConnectX()

print("INPUT  (Observation): ", env.observation_space)
print("OUTPUT (Agent action):", env.action_space)
```

```python
done = False
obs = env.reset()
while not done:
    # Choose first available empty column as the action.
    action = [i for i in range(len(obs.board)) if obs.board[i] == 0][0]
    obs, reward, done, info = env.step(action)
    print(obs)
env.render(mode="ipython")
```
